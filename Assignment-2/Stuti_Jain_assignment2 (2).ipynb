```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
def payoff_graph(k,p,option) :
    s = np.arange(k-500,k+500, 10)
    if p>=0:       #buying the option 
        if option.lower()== "call":
             Profit = np.maximum(0,s - k)-p
             plt.plot(s, Profit, label="Buy Call", color="mediumpurple")
        else :             #put option
             Profit = np.maximum(0, k-s)-p
             plt.plot(s, Profit, label="Buy Put", color="cornflowerblue")
    else :           
        #Selling the option
        P = -p
        if option.lower()== "call" :
            Profit = P - np.maximum(0,s -k)
            plt.plot(s, Profit, label="Sell Call", color="saddlebrown")
        else :              #put option
            Profit = P - np.maximum(0,k -s)
            plt.plot(s, Profit, label="Sell Put", color="midnightblue")
    plt.title("Options Payoff graphs")
    plt.xlabel("Spot Prices")
    plt.ylabel("Profit/Loss")
    plt.legend()
    plt.legend(loc="upper right")
    plt.grid(True)
    plt.axhline(0, color='black', linewidth=1)
    plt.show()
k = float(input("Enter the strike price"))
p = float(input("Enter the premium"))
option = input("Select a type of option: Call or Put")
payoff_graph(k,p,option)


#Explanation of all the greeks & their range for call and put option
# Delta - It is the change of price of premium based on the price of underlying asset's price.
     #It's range of Call option is 0-1, & for put option its (-1,0)
     #In deep ITM cases delta is approx 1 or -1, and in deep OTM cases delta is approx 0
#Gamma - It is the rate of change of delta, higher gamma shows high risk.
        #The value of gamma is always positive for long options and varies from 0-1
        #Gamma has highest values at ATM and approx 0 at deep ITM/OTM
#Theta - It is the measure of the decrease in premium's price with each passing day
         #It is negative for long options and positive for short options
#Vega - It is the measure of change in the option's price based on changes in implied volatility
        #Higher vega shows the market is  more volatile so the premium also rises
        #Both call and put options have positve vega
#Rho - It is the measure of change in options's price based on changes in risk free interest rates
       #Call options have positve rho while put options have negative rho
```
