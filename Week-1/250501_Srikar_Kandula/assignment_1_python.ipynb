```python
# Question 2

string = input("Enter a string: ")
palindrome = False

def check_palindrome(string,palindrome):
    length = len(string)
    index = 0
    while index < length//2:
        if string[index] == string[length-index-1]:
            palindrome = True
        else:
            palindrome = False
            break
        index += 1
    if palindrome == True:
        print('Palindrome')
    else:
        print('Not palindrome')


check_palindrome(string,palindrome)
```
